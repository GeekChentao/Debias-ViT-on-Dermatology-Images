```python
import pandas as pd
```

```python
concepts = [
    "Vesicle",
    "Papule",
    "Macule",
    "Plaque",
    "Abscess",
    "Pustule",
    "Bulla",
    "Patch",
    "Nodule",
    "Ulcer",
    "Crust",
    "Erosion",
    "Excoriation",
    "Atrophy",
    "Exudate",
    "Purpura/Petechiae",
    "Fissure",
    "Induration",
    "Xerosis",
    "Telangiectasia",
    "Scale",
    "Scar",
    "Friable",
    "Sclerosis",
    "Pedunculated",
    "Exophytic/Fungating",
    "Warty/Papillomatous",
    "Dome-shaped",
    "Flat topped",
    "Brown(Hyperpigmentation)",
    "Translucent",
    "White(Hypopigmentation)",
    "Purple",
    "Yellow",
    "Black",
    "Erythema",
    "Comedo",
    "Lichenification",
    "Blue",
    "Umbilicated",
    "Poikiloderma",
    "Salmon",
    "Wheal",
    "Acuminate",
    "Burrow",
    "Gray",
    "Pigmented",
    "Cyst",
]

df = pd.read_csv("train_data.csv")
concept_pick = 3

descriptions = list()

for idx, row in df.iterrows():
    description = "Light skin. " if int(row["skin_color"]) <= 4 else "Dark skin. "
    scores = [float(row[concept]) for concept in concepts]
    top_scores = sorted(enumerate(scores), key=lambda x: x[1], reverse=True)[
        :concept_pick
    ]
    for idx, score in top_scores:
        description += f"{concepts[idx]} "
    descriptions.append(description)

df["Most Present Concepts"] = descriptions
df.to_csv("train_data.csv", index=False)
```
